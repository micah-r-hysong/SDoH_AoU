# Make case-control df

```python
dfs = [asthma_df, afib_df, breast_cancer_df, kidney_df, 
       chd_df, hyper_df, prostate_cancer_df,
       final_t2d_df, final_t1d_df]

status_columns = ["Asthma", "Afib", "BreastC", "CKD", "CHD", "HyperC",  "ProstateC", "t2d", "t1d"]

# Extract and rename the status columns
status_dfs = []
for df, col_name in zip(dfs, status_columns):
        status_dfs.append(df[['person_id', 'status']].rename(columns={'status': col_name}))

# Merge all the status columns on 'person_id'
merged_df = status_dfs[0]
for status_df in status_dfs[1:]:
    merged_df = pd.merge(merged_df, status_df, on='person_id', how='outer')
```

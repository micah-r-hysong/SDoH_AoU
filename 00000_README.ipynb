# System requirements

## Operating system

The All of Us Research Program does not require researchers to use a specific local operating system (OS). Instead, all data analysis is conducted within a secure, cloud-based platform called the Researcher Workbench

## Software dependencies

R and Pyton

Python packages: pingouin 

R packages: psych, lavaan, caret, pROC, pheatmap

## Hardware requirements

No non-standard hardware required

Recommended:

≥ 15 GB RAM

≥ 4 CPU cores for parallel model fitting (optional)
